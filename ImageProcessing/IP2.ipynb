```python
import numpy as np
image_Org = cv2.imread('../image/Lena.png')

operated = image_Org.astype(np.int16)
operated += 60 # - add value
# operated = 255-operated :Inverse

operated_clipped = np.clip(operated, 0, 255).astype(np.uint8)

cv2.imshow("image _ clipped",operated_clipped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
original = cv2.imread("B:\muzik\pyhton\JupyterProject\image\CT_sample.tif", cv2.IMREAD_GRAYSCALE)
print(original)
```

```python
cv2.imshow("Original",original)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
hist = cv2.calcHist([original],[0],None,[256],[0,256])
x = np.arange(256)
y = hist.ravel()
plt.bar(x,y)
```

---

```python
equalizeds = cv2.equalizeHist(original)
hist = cv2.calcHist([equalizeds],[0],None,[256],[0,256])
x = np.arange(256)
y = hist.ravel()
plt.bar(x,y)
plt.show()

cv2.imshow("Original",equalizeds)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

tif {RGB}‡∏≠‡∏¢‡∏π‡πà‡πÅ‡∏•‡πâ‡∏ß ‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏ñ‡∏∂‡∏á‡∏Ñ‡πà‡∏≤‡πÑ‡∏´‡∏ô‡∏Å‡πá‡πÑ‡∏î‡πâ [:,:,0]

```python
streshed = cv2.normalize(original,None,alpha=0,beta=255,norm_type=cv2.NORM_MINMAX)
hist = cv2.calcHist([streshed],[0],None,[256],[0,256])
x = np.arange(256)
y = hist.ravel()
plt.bar(x,y)
plt.show()

cv2.imshow("Original",equalizeds)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
