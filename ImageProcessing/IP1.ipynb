```python
import cv2
import numpy as np
```

```python
image = cv2.imread("../image/Lena.png")
```

```python
print(image)
```

```python
image.shape
```

```python
cv2.imshow("Image",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
plt.grid(False)
plt.imshow(image,cmap="gray_r")
```

```python
imga_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.grid(False)
plt.imshow(imga_rgb)
```

---

```python
imga_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow("Gray Image",imga_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
print(imga_gray)
```

```python
imga_gray.shape
```

```python
# 1.  ( float)
# 2.  uint8 (Integer 0-255) 
image_maual_gray = (0.299*image[:,:,0] + 0.587*image[:,:,1] + 0.114*image[:,:,2]).astype(np.uint8)
```

```python
cv2.imshow("Gray Image",image_maual_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image_maual_gray = (0.35*image[:,:,0] + 0.55*image[:,:,1] + 0.18*image[:,:,2]).astype(np.uint8)
```

```python
cv2.imshow("Gray Image",image_maual_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
